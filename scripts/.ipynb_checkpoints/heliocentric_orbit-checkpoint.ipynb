```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def func_nbody(x, t):
    
    # constants
    GM = 1.327e+11 # heliocentric gravitational constant, km^3/s^2
    GM1 = 3.986e+5 # earth gravitational constants, km^3/s^2
    theta1 = np.deg2rad(44) # true anomaly, rad
    period1 =365*24*3600 # earth orbital period, seconds
    a1 = 149597070.7 # earth semi-major axis, km

    # earth position
    r1 = np.array([a1*np.cos(2*np.pi*t/period1 + theta1), a1*np.sin(2*np.pi*t/period1 + theta1)])
    
    # equation of motion
    d1 = x[0:2] - r1
    r_norm = np.sqrt(x[0]**2 + x[1]**2)
    d1_norm = np.sqrt(d1[0]**2 + d1[1]**2)
    
    if d1_norm < 6371:
        print("ERROR: Spacecraft's attitude is lower than the earth's radius")
        
    dxdt = [x[2], x[3], -GM*x[0]/(r_norm**3) - GM1*d1[0]/(d1_norm**3), -GM*x[1]/(r_norm**3) - GM1*d1[1]/(d1_norm**3)]
    
    return dxdt
            
#  initial conditions
x0 = np.array([1.05782e+8, 1.05782e+8, -21.06, 21.06]) # [position_x, position_y, velocity_x, velocity_y]
t = np.linspace(0, 365*24*3600, 100) # 1-year period of time split by 100 seconds

sol = odeint(func_nbody, x0, t)

# plot orbit
plt.plot(sol[:, 0],sol[:, 1], 'b')
plt.grid()
plt.gca().set_aspect('equal')
plt.xlabel('x, km')
plt.ylabel('y, km')
from astropy import units as u
from astropy.time import Time
plt.show()
```

```python
def func_nbody(x, t):
    
    # constants
    GM = 1.327e+11 # heliocentric gravitational constant, km^3/s^2
    GM1 = 3.986e+5 # earth gravitational constants, km^3/s^2
    theta1 = 0.0 # true anomaly, rad
    period1 =365*24*3600 # earth orbital period, seconds
    a1 = 149597070.7 # earth semi-major axis, km

    # earth position
    r1 = np.array([a1*np.cos(2*np.pi*t/period1 + theta1), a1*np.sin(2*np.pi*t/period1 + theta1)])
    
    # equation of motion
    d1 = x[0:2] - r1
    r_norm = np.sqrt(x[0]**2 + x[1]**2)
    d1_norm = np.sqrt(d1[0]**2 + d1[1]**2)
    
    if d1_norm < 6371:
        print("ERROR: Spacecraft's attitude is lower than the earth's radius")
        
    dxdt = [x[2], x[3], -GM*x[0]/(r_norm**3) - GM1*d1[0]/(d1_norm**3), -GM*x[1]/(r_norm**3) - GM1*d1[1]/(d1_norm**3)]
    
    return dxdt

# 条件の設定
r_earth = 149597870.7 # 地球の公転軌道半径, km
v_earth = np.sqrt(1.327e11/r_earth) # 地球の公転速度, km/s
vinf = 5 # 地球公転速度に対する相対速度V∞, km/s
v_sc_x = np.sqrt(4*v_earth**2-vinf**2)*vinf/(2*v_earth) # 探査機の公転速度のx成分, km/s
v_sc_y = (2*v_earth**2 - vinf**2)/(2*v_earth) # 探査機の公転速度のy成分, km/s

# 0年〜0.2年の間の軌道伝播（二体問題）
x0 = np.array([r_earth, 0.0, -v_sc_x, v_sc_y]) # 位置, 速度, km, km/s
t_span = np.linspace(0,0.2*365*24*60*60,200) # 0.2年分を200ステップで刻む
sol_0to1 = odeint(func_twobody, x0, t_span)

# 0.2年〜2年の間の軌道伝播（N体問題）
x1 = sol_0to1[-1,:] + [0,0,0.0055,0] # 軌道伝播された位置・速度に軌道制御(ΔV)を加える
t_span1 = np.linspace(0.2*365*24*60*60,2*365*24*60*60,200) # 0.2年〜2年分を200ステップで刻む
sol_1to2 = odeint(func_nbody, x1, t_span1)

# 軌道の描画
theta = np.linspace(0,2*np.pi,100)
plt.plot(r_earth*np.cos(theta),r_earth*np.sin(theta),'b', label="Earth") # 地球の軌道
plt.plot(sol_0to1[:, 0],sol_0to1[:, 1], 'r', label="Spacecraft Before Delta-V")
plt.plot(sol_1to2[:, 0],sol_1to2[:, 1], 'g', label="Spacecraft After Delta-V")
plt.grid() # 格子をつける
plt.legend(loc="lower left")
plt.gca().set_aspect('equal') # グラフのアスペクト比を揃える
plt.xlabel('x, km')
plt.ylabel('y, km')
plt.show()
```

```python
from astropy import units as u
from astropy.time import Time
from astropy.coordinates import solar_system_ephemeris

solar_system_ephemeris.set("jpl")

import numpy as np

from poliastro.bodies import Sun, Earth, Jupiter
from poliastro.ephem import Ephem
from poliastro.frames import Planes
from poliastro.maneuver import Maneuver
from poliastro.plotting import StaticOrbitPlotter
from poliastro.plotting import OrbitPlotter3D
from poliastro.twobody import Orbit
from poliastro.util import norm, time_range
from poliastro.twobody.sampling import TrueAnomalyBounds

# event dates
date_launch = Time("2011-08-05 16:25", format="iso", scale="utc").tdb
date_flyby =Time("2013-10-09 19:21", format="iso", scale="utc").tdb
date_arrival = Time("2016-07-05 03:18", format="iso", scale="utc").tdb

# Atras V supplied a launch energy
C_3 = 31.3 * u.km**2 / u.s**2

# # Plot initial Earth's position
# Earth.plot(date_launch, label="Initial Earth's position")

# Get Earth ephemeris at launch
earth = Ephem.from_body(Earth, time_range(date_launch, end=date_arrival, periods=500), plane=Planes.EARTH_ECLIPTIC)

# Get position and velocity vector at launch
r_e0, v_e0 = earth.rv(date_launch)

# Assume that the insertion velocity is tangential to that of the Earth
dv = C_3**0.5 * v_e0 / norm(v_e0)

# Create a delta-v maneuver from impulse constructor
man = Maneuver.impulse(dv)

ss_e0 = Orbit.from_ephem(Sun, earth, date_launch)
ss_efly = Orbit.from_ephem(Sun, earth, date_flyby)

# Inner Cruise 1
ic1 = ss_e0.apply_maneuver(man)
ic1_end = ic1.propagate_to_anomaly(180.0 * u.deg)

# # We can check new bodies positions
# plotter = StaticOrbitPlotter()
# plotter.plot_body_orbit(
#     Earth,
#     ic1_end.epoch,
#     label="Earth at end of flyby",
#     trail=True,
# )
# plotter.plot_maneuver(ss_e0, man, color="C1", label="Initial impulse")
# plotter.plot_trajectory(
#     ic1.to_ephem(strategy=TrueAnomalyBounds(min_nu=0 * u.deg, max_nu=180 * u.deg)).sample(),
#     label="Inner cruise 1",
#     color="C1",
# )

print("Inner cruise 1 period:", ic1.period.to(u.day))

# Let's compute the Lambert solution to do the flyby of the Earth
man_flyby = Maneuver.lambert(ic1_end, ss_efly)
imp_a, imp_b = man_flyby.impulses

# Solution of lambert arc
print("Initial impulse:", imp_a)
print("Final impulse:", imp_b)

# Check the initial delta-V
dv_a = imp_a[-1]
print("delta-V:", norm(dv_a.to(u.km / u.s)))

# Let us apply the maneuver
ic2, ss_flyby = ic1_end.apply_maneuver(man_flyby, intermediate=True)

# We propagate the transfer orbit till the flyby occurs
ic2_end = ic2.propagate(date_flyby)

# plotter = StaticOrbitPlotter()
# plotter.plot_body_orbit(
#     Earth,
#     ss_efly.epoch,
#     label="Earth's position at flyby",
#     trail=True,
# )
# plotter.plot_maneuver(ss_e0, man, color="C1", label="Initial impulse")
# plotter.plot_trajectory(
#     ic1.to_ephem(strategy=TrueAnomalyBounds(min_nu=0 * u.deg, max_nu=180 * u.deg)).sample(),
#     label="Inner cruise 1",
#     color="C1",
# )
# plotter.plot_maneuver(
#     ic1_end,
#     man_flyby,
#     label="Inner cruise 2 maneuver",
#     color="C2",
# )

# And now, go to Jupiter!
jupiter = Ephem.from_body(Jupiter, time_range(date_launch, end=date_arrival), plane=Planes.EARTH_ECLIPTIC)

ss_j = Orbit.from_ephem(Sun, jupiter, date_arrival)

# Solving for maneuver to Jupiter
man_jupiter = Maneuver.lambert(ic2_end, ss_j)
imp_a, imp_b = man_jupiter.impulses

print("Initial impulse:", imp_a)
print("Final impulse:", imp_b)

# Apply the maneuver
ic3, ss_target = ic2_end.apply_maneuver(man_jupiter, intermediate=True)
ic3_end = ic3.propagate(date_arrival)

# Final plot for the whole mission
plotter = StaticOrbitPlotter()
plotter.plot_body_orbit(
    Earth,
    date_flyby,
    label="Earth's position at flyby",
    trail=True,
)
plotter.plot_maneuver(ss_e0, man, color="C1", label="Initial impulse")
plotter.plot_trajectory(
    ic1.to_ephem(strategy=TrueAnomalyBounds(min_nu=0 * u.deg, max_nu=180 * u.deg)).sample(),
    label="Inner cruise 1",
    color="C1",
)
plotter.plot_maneuver(
    ic1_end,
    man_flyby,
    label="Inner cruise 2",
    color="C2",
)
plotter.plot_maneuver(
    ic2_end,
    man_jupiter,
    label="Inner cruise 3",
    color="C3",
)
plotter.plot_body_orbit(
    Jupiter, date_arrival, label="Jupiter's position at arrival", trail=True
)

plotter = OrbitPlotter3D(plane=Planes.EARTH_ECLIPTIC)
plotter.set_attractor(Sun)
plotter._layout=dict(scene=dict(aspectmode='manual', aspectratio=dict(x=1, y=1, z=1), xaxis=dict(range=[-1000e+6, 1000e+6]), yaxis=dict(range=[-1000e+6, 1000e+6]), zaxis=dict(range=[-1000e+6, 1000e+6]) ))
plotter.plot_body_orbit(
    Earth,
    date_flyby,
    label="Earth's position at flyby",
    trail=True,
)
plotter.plot_maneuver(ss_e0, man, color="orange", label="Initial impulse")
plotter.plot_trajectory(
    ic1.to_ephem(strategy=TrueAnomalyBounds(min_nu=0 * u.deg, max_nu=180 * u.deg)).sample(),
    label="Inner cruise 1",
    color="orange",
)
plotter.plot_maneuver(
    ic1_end,
    man_flyby,
    label="Inner cruise 2",
    color="green",
)
plotter.plot_maneuver(
    ic2_end,
    man_jupiter,
    label="Inner cruise 3",
    color="purple",
)
plotter.plot_body_orbit(
    Jupiter, date_arrival, label="Jupiter's position at arrival", trail=True
)
```

```python
# Let's reproduce the meneuver plan we just did in JAXA academy's work, this time using poliastro!
# To do so, we need to propagate the orbit by solving the N-body problem using poliastro.

from astropy import units as u
from astropy.time import Time
from astropy.coordinates import solar_system_ephemeris

solar_system_ephemeris.set("jpl")

import numpy as np
import cv2

from poliastro.bodies import Sun, Earth, Jupiter
from poliastro.ephem import Ephem
from poliastro.frames import Planes
from poliastro.maneuver import Maneuver
from poliastro.plotting import StaticOrbitPlotter
from poliastro.plotting import OrbitPlotter3D
from poliastro.twobody import Orbit
from poliastro.util import norm, time_range
from poliastro.twobody.sampling import TrueAnomalyBounds
from poliastro.twobody.propagation import (
    ALL_PROPAGATORS,
    ELLIPTIC_PROPAGATORS,
    HYPERBOLIC_PROPAGATORS,
    PARABOLIC_PROPAGATORS,
    CowellPropagator,
    DanbyPropagator,
    FarnocchiaPropagator,
    GoodingPropagator,
    MarkleyPropagator,
    RecseriesPropagator,
    ValladoPropagator,
)

# event dates
t0 = Time("2022-09-23 00:00", format="iso", scale="utc")
t1 = Time(Time(t0.jd+365, format="jd", scale="utc").iso, format="iso", scale="utc")
t2 = Time(Time(t1.jd+365, format="jd", scale="utc").iso, format="iso", scale="utc")
date_launch = t0.tdb
date_flyby =t1.tdb
date_arrival = t2.tdb

# get Earth ephemeris at launch
earth = Ephem.from_body(Earth, time_range(date_launch, end=date_arrival, periods=500), plane=Planes.EARTH_ECLIPTIC)

# get position and velocity vector at launch
r_e0, v_e0 = earth.rv(date_launch)
v_e0 = v_e0 / 24 / 3600 * u.d / u.s

# launch energy (Vinf = 5 km/s)
C_3 = 5.0**2 * u.km**2 / u.s**2

# calculate position and velocity vectors to put spacecraft in 1:1 resonant orbit
nr_e0 = r_e0 / np.linalg.norm(r_e0)
nv_e0 = v_e0 / np.linalg.norm(v_e0)
theta = 2 * np.arcsin(C_3.value**0.5 / 2 / np.linalg.norm(v_e0.value))
rvec = theta * np.cross(nr_e0, nv_e0)
rmat, _ = cv2.Rodrigues(rvec)
r_sc0 = r_e0 # position at launch
v_sc0 = np.dot(rmat, v_e0) # velocity at launch

# create an initial delta-v maneuver at launch
dv = v_sc0 - v_e0
man = Maneuver.impulse(dv)

# spacecraft orbit before launch
sc_e0 = Orbit.from_ephem(Sun, earth, date_launch)

# inner cruise 1
ic1 = sc_e0.apply_maneuver(man)
ic1_end = ic1.propagate_to_anomaly(0 * u.deg)
min_nu_ic1 = np.rad2deg(ic1.nu.value) * u.deg # true anomaly at launch
max_nu_ic1 = 0.0 * u.deg # perihelion

print("Earth velocity at launch: ", v_e0)
print("Spacecraft velocity at launch: ", v_sc0)

# create a deep space delta-v maneuver at the end of inner cruise 1 = perihelion
dv_peri = (0.0055 * u.km / u.s) * ic1_end.v / np.linalg.norm(ic1_end.v)
man_peri = Maneuver.impulse(dv_peri)

# inner cruise 2 with deep space maneuver needed for Earth flyby
ic2 = ic1_end.apply_maneuver(man_peri)

# How to compute the Earth Flyby using poliastro?
# (0) Compute the "date_flyby"
# (1) Propagate ic2 until "date_flyby"
# (2) Analyze Earth flyby and compute escape velocity
# (3) apply the result of Earth flyby as man_flyby at "date_flyby"

date_flyby = ic2.epoch
for i in range(100):
    tof = i * u.d 
    date_flyby += tof
    ic2_end = ic2.propagate(date_flyby)
    ic2_end.r.value

plotter = OrbitPlotter3D(plane=Planes.EARTH_ECLIPTIC)
plotter.set_attractor(Sun)
plotter.plot_body_orbit(Earth, date_flyby, label="Earth's position at launch")
plotter.plot_maneuver(ic1, man, color="orange", label="Initial impulse (%.1f km / s)" % np.linalg.norm(dv.value))
plotter.plot_trajectory(
    ic1.to_ephem(strategy=TrueAnomalyBounds(min_nu_ic1, max_nu_ic1)).sample(),
    label="Inner cruise 1",
    color="orange",
)
plotter.plot_maneuver(ic2, man_peri, color="purple", label="Deep space impulse (%.1f km / s)" % np.linalg.norm(dv_flyby.value))
plotter.plot_trajectory(
    ic2.to_ephem(strategy=TrueAnomalyBounds(max_nu_ic1, max_nu_ic1+180*u.deg)).sample(),
    label="Inner cruise 2",
    color="purple",
)
```
